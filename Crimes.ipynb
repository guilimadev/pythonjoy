```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```
url_il = 'https://raw.githubusercontent.com/guilimadev/pythonjoy/master/csv/IL_CrimeStatebyState.csv'

url_ny = 'https://raw.githubusercontent.com/guilimadev/pythonjoy/master/csv/NY_CrimeStatebyState.csv'

df_il = pd.read_csv(url_il, index_col=0)
df_il.fillna(0,inplace=True)
print(df_il.tail())

df_ny = pd.read_csv(url_ny, index_col=0)
df_ny.fillna(0,inplace=True)
print(df_ny.tail())
```

```
columns = df_ny.columns

columns
```

```
df_il2 = df_il[columns]
df_ny.head()
```

```
df_ny.plot(y=['Violent crime total', 'Property crime total'], grid=True)
plt.title('NY - 1965 to 2014')
plt.show()

df_il2.plot(y=['Violent crime total', 'Property crime total'], grid=True)
plt.title('Illinois - 1965 to 2014')
plt.show()
```

```
df_ny.plot(y=['Violent crime total', 'Robbery'], grid=True)
plt.title('NY')
plt.show()

df_il2.plot(y=['Violent crime total', 'Robbery'], grid=True)
plt.title('Illinois')
plt.show()
```

```
print('Interpretation:' + '\n' + 'The correlation coefficient ranges from −1 to 1. A value of 1 implies that a linear equation describes the relationship between X and Y perfectly, with all data points lying on a line for which Y increases as X increases.' + '\n' + 'A value of −1 implies that all data points lie on a line for which Y decreases as X increases. A value of 0 implies that there is no linear correlation between the variables.')

correlation_il = np.corrcoef(df_il2['Robbery'], df_il2['Violent crime total'])

print('\n'+ 'State of Illinois correlation of Robbery and Violent crime total: ' + str(correlation_il[0][1]))

correlation_ny = np.corrcoef(df_ny['Robbery'], df_ny['Violent crime total'])

print('State of New York correlation of Robbery and Violent crime total: ' + str(correlation_ny[0][1]))
```
